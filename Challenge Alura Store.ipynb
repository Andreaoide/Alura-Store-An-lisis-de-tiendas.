### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



**1.Ingreso total de cada tienda**

```python
ingresos_tienda1 = tienda1["Precio"].sum()
ingresos_tienda2 = tienda2["Precio"].sum()
ingresos_tienda3 = tienda3["Precio"].sum()
ingresos_tienda4 = tienda4["Precio"].sum()

```

```python
print("Ingreso Total Tienda 1:", ingresos_tienda1)
print("Ingreso Total Tienda 2:", ingresos_tienda2)
print("Ingreso Total Tienda 3:", ingresos_tienda3)
print("Ingreso Total Tienda 4:", ingresos_tienda4)
```

# 2. Ventas por categoría

```python
ventas_categoria_tienda1 = tienda1["Categoría del Producto"].value_counts()
ventas_categoria_tienda2 = tienda2["Categoría del Producto"].value_counts()
ventas_categoria_tienda3 = tienda3["Categoría del Producto"].value_counts()
ventas_categoria_tienda4 = tienda4["Categoría del Producto"].value_counts()
```

# 6. Comparación de tiendas


```python
facturacion = {
    "Tienda 1": tienda1["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

calificaciones = {
    "Tienda 1": tienda1["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}

envio_promedio = {
    "Tienda 1": tienda1["Costo de envío"].mean(),
    "Tienda 2": tienda2["Costo de envío"].mean(),
    "Tienda 3": tienda3["Costo de envío"].mean(),
    "Tienda 4": tienda4["Costo de envío"].mean()
}
```

```python
resumen = pd.DataFrame({
    "Facturación": facturacion,
    "Calificación Promedio": calificaciones,
    "Costo Envío Promedio": envio_promedio
})
resumen.sort_values(by="Facturación", ascending=False)
resumen = resumen.round(2)
resumen
```

**Objetivos**



El objetivo principal de este proyecto consiste en comparar cuatro tiendas en cuanto a 4 variables principales:
Los ingresos totales de las tiendas.
Las categorías de productos más y menos vendidas.
Las calificaciones promedio de los clientes por tienda.
Los productos más y menos vendidos.
El coste de envío promedio para cada tienda.
Esto con el fin de hacer un análisis detallado para comparar las 4 tiendas y ayudar al señor Juan a tomar una decisión más informada sobre que tienda le convendría vender.

**Resultados**


El análisis de los datos de las cuatro tiendas permitió identificar varias diferencias importantes en su desempeño comercial y en la experiencia de los clientes.

1.	Facturación
En primer lugar, el análisis de ingresos totales muestra que la Tienda 1 presenta la mayor facturación, con un total de 1,150,880,400, seguida por la Tienda 2 y la Tienda 3. Por su parte, la Tienda 4 presenta la menor facturación, lo que indica un menor volumen de ventas en comparación con las demás tiendas.

2.	Venta de productos específicos y por categoría
Asimismo, el análisis de los productos más y menos vendidos en cada tienda permitió identificar qué artículos tienen mayor aceptación entre los clientes y cuáles presentan menor demanda. En la Tienda 1, el producto más vendido es el Televisor LED UHD 4K, mientras que el Microondas aparece entre los productos con menor número de ventas. En la Tienda 2, el producto más vendido es la Cámara digital, mientras que el Juego de mesa presenta uno de los niveles de ventas más bajos. En la Tienda 3, el Kit de bancas destaca como el producto con mayor número de ventas, mientras que la Guitarra eléctrica aparece entre los menos vendidos. Finalmente, en la Tienda 4, el producto más vendido es la Cama box, mientras que el Guitarra eléctrica presenta una menor cantidad de ventas.
En relación con las categorías de productos, el análisis conjunto de las cuatro tiendas muestra que las categorías con mayor número de ventas son Muebles (465 ventas) y Electrónicos (448 ventas), lo que indica que estos productos concentran gran parte de la demanda de los clientes. En contraste, las categorías con menor número de ventas son Artículos para el hogar (171 ventas), Libros (173 ventas) e Instrumentos musicales (182 ventas).

3.	Calificación de satisfacción de los clientes
En cuanto a la satisfacción de los clientes, se observa que la Tienda 3 presenta la calificación promedio más alta (4.05), seguida por la Tienda 2 con 4.04, mientras que la Tienda 1 presenta la calificación promedio más baja (3.98).

4.	Costos promedio de envío
Por último, el análisis del costo promedio de envío muestra que la Tienda 1 tiene el costo de envío promedio más alto (26018.61), mientras que la Tienda 4 presenta el costo promedio más bajo (23459.46).


**Conclusión**

A partir del análisis de los datos de las cuatro tiendas, se identificaron diferencias relevantes en términos de ingresos, satisfacción de los clientes, categorías de productos y costos logísticos.

El análisis de ingresos totales muestra que la Tienda 1 genera la mayor facturación, seguida por la Tienda 2 y la Tienda 3. En contraste, la Tienda 4 presenta la menor facturación, lo que indica un menor volumen de ventas en comparación con las demás tiendas.

En cuanto a las categorías de productos, los resultados muestran que Muebles y Electrónicos son las categorías con mayor número de ventas en el conjunto de las tiendas, lo que indica que estos productos concentran la mayor parte de la demanda. Por otro lado, categorías como Artículos para el hogar, Libros e Instrumentos musicales presentan una menor participación en las ventas.

El análisis de productos más y menos vendidos permitió identificar qué artículos tienen mayor aceptación entre los clientes, evidenciando que algunos productos relacionados con muebles y tecnología presentan una mayor demanda, mientras que otros productos tienen menor rotación dentro de las tiendas.

En relación con la satisfacción de los clientes, la Tienda 3 presenta la mejor calificación promedio, lo que sugiere una experiencia de compra más positiva en comparación con las otras tiendas. Por otro lado, la Tienda 1 presenta la calificación promedio más baja, lo que podría indicar oportunidades de mejora en la experiencia del cliente.

Finalmente, el análisis del costo promedio de envío muestra que la Tienda 1 tiene el costo de envío más alto, mientras que la Tienda 4 presenta el costo de envío más bajo.

Considerando en conjunto todos los factores analizados, se recomienda que el Sr. Juan venda la Tienda 4. Aunque esta tienda presenta costos logísticos más bajos, también es la que genera la menor facturación dentro del negocio, lo que indica que su impacto en los ingresos totales es menor en comparación con las otras tiendas.

Desde una perspectiva estratégica, mantener las tiendas con mayor volumen de ventas permite conservar los activos que generan más ingresos para el negocio, mientras que vender la tienda con menor desempeño financiero puede representar una oportunidad para optimizar recursos y concentrar esfuerzos en las tiendas con mejor rendimiento.
